```python
# Importing Libraries
import pandas as pd
```

```python
# Loading and Displaying Data
df = pd.read_csv("../../Data/Rawdata/uszips.csv")
df.head()
```

```python
# Filtering Data by State
df = df.loc[df['state_id'] == 'IL']
df
```

```python
# Selecting and Renaming Columns
df = df[['zip', 'lat', 'lng']]
df.columns = ['zip_code', 'latitude', 'longitude']
df
```

```python
# Resetting the Index
df = df.reset_index(drop=True)
df
```

```python
# Removing Missing Values
df = df.dropna(how="any")
df
```

```python
# Saving the Cleaned Data
df.to_csv('../../Data/CleanData/illinois.csv', index = False)
```
